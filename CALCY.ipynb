```python
import tkinter as tk
from tkinter import *
from tkinter import messagebox
from tkinter.font import Font
import sys
import math

root= Tk()
root.title("Simple Calculator")
root.iconbitmap(r'C:\Users\DinAnsh\Downloads\project_zero\calcy.ico')
e = Entry(root, width=23, borderwidth=1.50, font=Font(family='Times',size='14'))
e.grid(row=0, column=0, columnspan=3, pady=1.50, padx=1.50)

def button_click(number):
    e.insert(len(e.get())+1,number)

def clear():
    e.delete(0,END)

def bkspc():
    e.delete(len(e.get())-1,END)
    
def button_addition():
    e.insert(len(e.get())+1,"+")

def button_subtract():
    e.insert(len(e.get())+1,"-")

def button_multiply():
    e.insert(len(e.get())+1,"*")

def button_divide():
    e.insert(len(e.get())+1,"/")

def dot():
    e.insert(len(e.get())+1,".")
    
def button_modulo():
    e.insert(len(e.get())+1,"%")

def last_op(expr):
    op = ['+','-','*',')','/',]
    exp = ''
    for e in expr[::-1]:
        if e in op:
            break
        else:
            exp+=e
    return len(exp)

def square():
    e.insert(len(e.get())-last_op(e.get()),'sqr(')
    e.insert(len(e.get())+last_op(e.get()),')')

def sqr_root():
    e.insert(len(e.get())-last_op(e.get()),'sqrt(')
    e.insert(len(e.get())+last_op(e.get()),')')
    
def fn_pow():
    e.insert(len(e.get())+last_op(e.get()),'^')

def fn_fact():
    e.insert(len(e.get())-last_op(e.get()),'fact(')
    e.insert(len(e.get())+last_op(e.get()),')')

def fn_ln():
    e.insert(len(e.get())-last_op(e.get()),'ln(')
    e.insert(len(e.get())+last_op(e.get()),')')  

def fn_log():
    e.insert(len(e.get())-last_op(e.get()),'log(')
    e.insert(len(e.get())+last_op(e.get()),')')

#need improvement
def pie():
    num1 = math.pi
    e.insert(1,num1)
    
def euler():
    num1 = math.e
    e.insert(1,num1)

def button_equal():
    exp = e.get()
    exp.replace('^','**')
    e.delete(0,END)
    try:
        exp = eval(exp)
    except:
        exp = sys.exc_info()[1].args[0]
    e.insert(1,exp)

def sqrt(num):
    return math.sqrt(num)
def fact(num):
    return math.factorial(num)
def ln(num):
    return math.log(num)
def log(num):
    return math.log10(num)
def sqr(num):
    return num**2

#Creating Buttons 
button_clear = Button(root,text="CLR",padx=17,pady=14,command=clear)
button_bkspc = Button(root,text='⇽',padx=17,pady=14,command=bkspc)

button_1 = Button(root,text="1",padx=25,pady=25,command=lambda : button_click(1))
button_2 = Button(root,text="2",padx=25,pady=25,command=lambda : button_click(2))
button_3 = Button(root,text="3",padx=25,pady=25,command=lambda : button_click(3))
button_4 = Button(root,text="4",padx=25,pady=25,command=lambda : button_click(4))
button_5 = Button(root,text="5",padx=25,pady=25,command=lambda : button_click(5))
button_6 = Button(root,text="6",padx=25,pady=25,command=lambda : button_click(6))
button_7 = Button(root,text="7",padx=25,pady=25,command=lambda : button_click(7))
button_8 = Button(root,text="8",padx=25,pady=25,command=lambda : button_click(8))
button_9 = Button(root,text="9",padx=25,pady=25,command=lambda : button_click(9))
button_0 = Button(root,text="0",padx=25,pady=25,command=lambda : button_click(0))

button_sqr = Button(root,text="x²",padx=25,pady=25,command=square)
button_root = Button(root,text="√x",padx=25,pady=25,command=sqr_root)
button_fact = Button(root,text="x!",padx=25,pady=25,command=fn_fact)
button_pie = Button(root,text="π",padx=25,pady=25,command=pie)
button_e = Button(root,text="e",padx=25,pady=25,command=euler)
button_ln = Button(root,text="ln",padx=25,pady=25,command=fn_ln)
button_log = Button(root,text="log",padx=20,pady=25,command=fn_log)
button_pow = Button(root,text="xⁿ",padx=25,pady=25,command=fn_pow)


button_div = Button(root,text="/",padx=25,pady=25,command=button_divide)
button_mul = Button(root,text="*",padx=25,pady=25,command=button_multiply)
button_min = Button(root,text="-",padx=25,pady=25,command=button_subtract)
button_plus = Button(root,text="+",padx=25,pady=25,command=button_addition)
button_eq = Button(root,text="=",padx=25,pady=25,command= button_equal)
button_dot = Button(root,text=".",padx=25,pady=25,command=dot)
button_mod = Button(root,text="%",padx=25,pady=25,command=button_modulo)

#Placing buttons in grid
button_clear.grid(row=0,column=4,sticky=tk.E+tk.W)
button_bkspc.grid(row=0,column=3,sticky=tk.E+tk.W)

button_sqr.grid(row=1,column=0,sticky=tk.E+tk.W)
button_root.grid(row=1,column=1,sticky=tk.E+tk.W)
button_fact.grid(row=1,column=2,sticky=tk.E+tk.W)
button_pie.grid(row=1,column=3,sticky=tk.E+tk.W)
button_e.grid(row=1,column=4,sticky=tk.E+tk.W)

button_7.grid(row=2,column=0,sticky=tk.E+tk.W)
button_8.grid(row=2,column=1,sticky=tk.E+tk.W)
button_9.grid(row=2,column=2,sticky=tk.E+tk.W)
button_ln.grid(row=2,column=3,sticky=tk.E+tk.W)
button_log.grid(row=2,column=4,sticky=tk.E+tk.W)

button_4.grid(row=3,column=0,sticky=tk.E+tk.W)
button_5.grid(row=3,column=1,sticky=tk.E+tk.W)
button_6.grid(row=3,column=2,sticky=tk.E+tk.W)
button_mod.grid(row=3,column=3,sticky=tk.E+tk.W)
button_pow.grid(row=3,column=4,sticky=tk.E+tk.W)

button_1.grid(row=4,column=0,sticky=tk.E+tk.W)
button_2.grid(row=4,column=1,sticky=tk.E+tk.W)
button_3.grid(row=4,column=2,sticky=tk.E+tk.W)
button_mul.grid(row=4,column=3,sticky=tk.E+tk.W)
button_div.grid(row=4,column=4,sticky=tk.E+tk.W)

button_dot.grid(row=5,column=0,sticky=tk.E+tk.W)
button_0.grid(row=5,column=1,sticky=tk.E+tk.W)
button_eq.grid(row=5,column=2,sticky=tk.E+tk.W)
button_min.grid(row=5,column=3,sticky=tk.E+tk.W)
button_plus.grid(row=5,column=4,sticky=tk.E+tk.W)

root.mainloop()
```
